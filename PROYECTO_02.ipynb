Creamos la función que genere las nuevas columnas

```python
def operacion(calculo: str='calculo1',a: float=100, b: float=100 )-> float:
  
  ''' 
  Retorna las siguientes operaciones dependiendo de "calculo":
  calculo='calculo1': log10(a/b) ,
  calculo='calculo2': sqrt(a)*np.exp(b)/200 ,
  calculo='calculo3': where(a>b,5,3),
  calculo='calculo4': 100/np.logn(a/b),
  calculo='calculo5': b**2/a
  Parámetros
  ----------
  'calculo' : variable de tipo str
  `a` : variable de tipo float.
  `b` : variable de tipo float.
 
  Ejemplos:
  --------
  calculo='calculo1'
  a=4
  b=2
  >>> operacion(calculo='calculo5')(a=2,b=3) # ingresando nuestros argumentos
  >>> 4.5

  '''
  return {
      'calculo1': lambda a, b: np.log10(a/b) ,
      'calculo2': lambda a, b: np.sqrt(a)*np.exp(b)/200 ,
      'calculo3': lambda a, b: np.where(a>b,5,3),
      'calculo4': lambda a, b: 100/np.log(a/b),
      'calculo5': lambda a, b: b**2/a
  }.get(calculo, lambda: None)
  
```

Aplicando la función operación

```python
data_autos["columna1"]=operacion(calculo='calculo1')(a=data_autos["Millas_Por_Galon"],b=data_autos["Caballo_De_Fuerza"])   
data_autos["columna2"]=operacion(calculo='calculo2')(a=data_autos["Millas_Por_Galon"],b=data_autos["Caballo_De_Fuerza"])   
data_autos["columna3"]=operacion(calculo='calculo3')(a=data_autos["Millas_Por_Galon"],b=data_autos["Caballo_De_Fuerza"])   
data_autos["columna4"]=operacion(calculo='calculo4')(a=data_autos["Millas_Por_Galon"],b=data_autos["Caballo_De_Fuerza"])   
data_autos["columna5"]=operacion(calculo='calculo5')(a=data_autos["Millas_Por_Galon"],b=data_autos["Caballo_De_Fuerza"])
data_autos  
```

Generamos la función

```python
def operacionRango(column: str, df:pd.DataFrame)-> float:
  
  ''' 
  Retorna la siguiente condición:
  "Alto":   si el valor de la columna es mayor que mean+std
  "Medio":  si el valor de la columna está entre mean-std y mean+std
  "Bajo":   si el valor de la columna es menor que mean-std
  
  Parámetros
  ----------
  `column` : variable de tipo str.
  `df` : variable de tipo DataFrame.
 

  '''

  std=np.std(df[column])
  mean=np.mean(df[column])
  high=std+mean
  low=std-mean
  return df[column].apply(lambda x: 
                              "Alto" if x>high else (
                                                       "Bajo" if x<low else (
                                                            "Medio"                                                        
                                                                            )
                                                    )
                      )

```

```python
def graficos_cajas(columnas: list, data: pd.DataFrame, x: str): # parámetros
  '''
  Múltiples gráficos de variables numéricas con una variable cualitativa.

  Parámetros
  ----------
  
  `columnas` : lista con los nombres de las variables numéricas.
  `data` : de tipo DataFrame.
  `x` : variable cualitativa | variable de exploración.
  
  Esta función graficará 4 gráficos para cada variable numérica ingresada 
  en la `lista`, según la variable `x`. Los gráficos son:
  - Gráfico de caja
  - Histograma
  - Barras
  -Histograma con curva 

  Ejemplos:
  --------

  '''
  
  for i,column in enumerate(columnas): 
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))
    sns.boxplot( x = x,y=column,  data = data , ax=axes[0])
    sns.histplot(x=x,  data = data, ax=axes[1])
    sns.barplot(x =x , y =column , estimator = np.mean, data=data,ax=axes[2])
    axes[1].set_title(column, fontsize = 10, fontweight = "bold")
    plt.figure();
    sns.displot(data=data, x=x, kde=True)
  #  print(f" ", column)
graficos_cajas(columnas=columnas_numericas,data=data_autos,x='Pais_Origen')   
```

## **EJERCICIO N° 05**

Definimos la función a usar

```python
def graficos_correlacion(columna: str, data: pd.DataFrame, x: str, fig:int=1): # parámetros
  '''
  Gráfico de un mapa de calor de múltiples variables numéricas con una variable cualitativa.

  Parámetros
  ----------
  
  `columna` : De tipo str
  `data` : de tipo DataFrame.
  `x` : variable cualitativa | variable de exploración.
  
  Esta función graficará la correlación entre variables numéricas basándose en
   una variable cualitativa de una determinada columna.


  '''
  plt.figure(num=fig,figsize=(8,6))
  data_correlacionar = data[data[columna]==x]
 

  correlacion = data_correlacionar.corr() # pearson

  p=sns.heatmap(correlacion,
            xticklabels = correlacion.columns.values,
            yticklabels = correlacion.columns.values,
            annot = True,
            cmap = "rocket"  # viridis, Blues
            )
  p.set_title(x)  

```

Graficamos separando la data por país

```python

graficos_correlacion(columna="Pais_Origen", data= data_autos, x= "Europe",fig=1)
graficos_correlacion(columna="Pais_Origen", data= data_autos, x= "USA",fig=2)
graficos_correlacion(columna="Pais_Origen", data= data_autos, x= "Japan",fig=3)
```
